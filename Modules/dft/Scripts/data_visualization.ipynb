```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# loading the dataset for training models
original_dataset_df = pd.read_csv('../Features Data/real_videos_features_vector.csv', index_col=None)
fake_dataset_1_df = pd.read_csv('../Features Data/fake_videos_features_vector_0.csv', index_col=None)
fake_dataset_2_df = pd.read_csv('../Features Data/fake_videos_features_vector_1.csv', index_col=None)
fake_dataset_3_df = pd.read_csv('../Features Data/fake_videos_features_vector_2.csv', index_col=None)
fake_dataset_4_df = pd.read_csv('../Features Data/fake_videos_features_vector_3.csv', index_col=None)
```

```python
# preprocessing
original_dataset_df.drop('Unnamed: 0', inplace=True, axis=1)
fake_dataset_1_df.drop('Unnamed: 0', inplace=True, axis=1)
fake_dataset_2_df.drop('Unnamed: 0', inplace=True, axis=1)
fake_dataset_3_df.drop('Unnamed: 0', inplace=True, axis=1)
fake_dataset_4_df.drop('Unnamed: 0', inplace=True, axis=1)
fake_dataset_1_df['Label'] = 1
fake_dataset_2_df['Label'] = 1
fake_dataset_3_df['Label'] = 1
fake_dataset_4_df['Label'] = 1
```

### Model 1

```python
# original dataset mean and std
dataset_1_real_mean = original_dataset_df.drop(['Label'], axis=1).mean(axis=0).to_numpy()
dataset_1_real_std = original_dataset_df.drop(['Label'], axis=1).std(axis=0).to_numpy()
```

```python
# fake dataset mean and std
dataset_1_fake_mean = fake_dataset_1_df.drop(['Label'], axis=1).mean(axis=0).to_numpy()
dataset_1_fake_std = fake_dataset_1_df.drop(['Label'], axis=1).std(axis=0).to_numpy()
```

```python
x = np.arange(len(dataset_1_real_mean))
fig, ax = plt.subplots()

ax.plot(x, dataset_1_real_mean, alpha=0.5, color='red', label='Real')
ax.fill_between(x, dataset_1_real_mean - dataset_1_real_std, dataset_1_real_mean + dataset_1_real_std, color='red', alpha=0.2)

ax.plot(x, dataset_1_fake_mean, alpha=0.5, color='blue', label='Fake')
ax.fill_between(x, dataset_1_fake_mean - dataset_1_fake_std, dataset_1_fake_mean + dataset_1_fake_std, color='blue', alpha=0.2)

ax.legend()
plt.title('Deepfake dataset')
plt.xlabel("Spatial Frequency")
plt.ylabel("Power Spectrum")
```
